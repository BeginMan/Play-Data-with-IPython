## 7. 該認真的來資料分析了

### 7-1 概念

用數學的角度來說, 資料分析大概可以看成「找函數」的工作。

    f(x) = y

#### [例子]

輸入一個時間 x, 我就出現那個時間花蓮市的溫度。

    f(x) = 在 x 時花蓮的溫度

#### [例子]

某公司做客戶分析, 把客戶分成三類:

* 1: 不用花什麼力氣就是我們忠實粉絲
* 2: 有可能成為客戶但要努力說服
* 3: 別傻了, 這人不會理我們的

我們自然希望把某一個人 x 放進我們的神秘函數, 函數就告訴我們他是哪一類的人:

    f(x) = x 這個人是哪一類
    
很棒吧? 等等, x 這個人電腦怎麼「輸入」? 所以我們會用某些可量化數據方式表式, 例如:

    x = {'血型':'O', '星座':'射手座', 'A產品':27, 
    'B產品':0, ...}
    
耶, 血型星座不是數字啊, 但我們可以用 1 表 A 型, 2 表 B 型, 等等。所以 x 總結可能是一個 list:

    x = [3, 1, 27, 0, ...]

#### [例子]

於是我們知道, 人可以「輸入」後, 世界又更寛廣了。比方說我們還可以做個「配對函數」, 輸入 x, y 兩個人, 得到:

    f(x,y) = x 和 y 的素配指數

#### [型式]

基本上我們學函數大約有兩種方式:

1. regression (迴歸)
2. interpolation (插值)


### 7-2 做個假的資料準備迴歸一下

我們來一條線, 比如說

    f(x) = 1.2x + 0.8

準備好個 50 個點

```python
%pylab inline
x = linspace(0,5,50)
```

對應的 y　就是我們要學的, 但是你弄得 y 都剛好是 $1.2x + 0.8$ 的話...

怎麼學都是 100% 像啊。

所以, 就像真實的世界, 我們加入 noise。

```python
y = 1.2*x + 0.8 + randn(50)
```

看看我們目前取的點, 附上「完全正確的」函數。

```python
scatter(x,y)
plot(x,1.2*x + 0.8,'b');
```

### 7-3 用 SciPy 來迴歸

線性迴歸很多套件會做, 我們這裡介紹的方法好處是你只要預想好這些資料長得很像什麼函數, 我們都可以迴歸! 不一定要線性函數!

在我們線性的例子很容易。我們先做一個目標函數

    f(x) = ax + b

再來算 a, b 這樣。

```python
from scipy.optimize import curve_fit
```

```python
def f(x, a, b):
    return a*x + b
```

我們「學習的資料」就是剛剛的 x, y。傳回的 popt 這裡就是最小平方法下得到的 a, b。

```python
popt, pcov = curve_fit(f, x, y)
```

```python
popt
```

```python
a, b = popt
```

```python
scatter(x,y)
plot(x,1.2*x + 0.8,'b',label="original")
plot(x, a*x + b, 'r',label="fit")
ax = gca()
ax.legend();
```

### 7-4 再來準備資料給插值法

插值法的目標是全部命中!!

### 7-5 全部命中其實不是什麼了不起的事

### 8-1 SVM 支持向量機

SVM (Support Vector Machines) 是用直線、曲線、平面、曲面等等把各類資料分開的方式。

來個很簡單的分類。

#### [例子]

第一類:
(-1, -1), (-2, -1)

第二類:
(1, 1), (2, 1)

這樣告訴 Python 什麼樣的點是哪一類。

```python
X = array([[-1, -1], [-2, -1], [1, 1], [2, 1]])
Y = array([1, 1, 2, 2])
```

### 8-2 使用 scikit-learn (sklearn)

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(X,Y)
```

用我們學成的函數預測新的資料

```python
clf.predict([[-0.8,-1]])
```

原始資料畫出來

```python
scatter(X[:,0], X[:,1],s=50,c=Y);
```

放一堆資料讓它分類看看

```python
gd = array([[i,j] for i in arange(-2.5, 2.5, 0.2) 
 for j in arange(-2.5, 2.5, 0.2)])
```

```python
gdc = clf.predict(gd)
```

```python
scatter(gd[:,0], gd[:,1], s=50, c=gdc);
```

### 8-3 生出假的資料給 SVM

剛剛的例子太簡單, 我們照例來生生假的訓練資料。sciket-leran 居然也想到要幫我們做這件事。

平面上隨便抽點分三類。

```python
X, Y = make_classification(n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, n_classes=3)
```

### 8-4 K-Mean 是自動分類

所謂非監督式學習就是「沒有標準解答」的學習。比如說們處理一些數據, 本來沒有分類, 讓電腦自動分類。

K-mean 是這種的方法。

```python
X = randn(50,2)
```

```python
scatter(X[:,0], X[:,1], s=55);
```

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters=3)
```

```python
clf.fit(X)
```

```python
clf.labels_
```

```python
scatter(X[:,0], X[:,1], s=55, c=clf.labels_)
```

```python
gdc = clf.predict(gd)
scatter(gd[:,0], gd[:,1], s=50, c=gdc);
```
